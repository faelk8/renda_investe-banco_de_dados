```python
def ativo_id():
    """
    Retorna o próximo ID para a tabela demonstrativo_resultado.ativo_total.
    O ID é gerado com base no maior valor já existente na tabela.
    Se a tabela estiver vazia, o ID começará em 1.
    """
    df = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/demonstrativo_resultado.csv')
    if df.empty:
        return 1
    else:
        return df['ativo_id'].max() + 1

def inserir_ativo(base, lista_prefixos, lista_sufixos, lista_valores, ID):
    # con = PostgresConnector()
    for nome, descricao, valor in zip(lista_prefixos, lista_sufixos, lista_valores):
        try:
            
            if base == "Despesas/Receitas Operacionais":
                temp = pd.DataFrame(columns=['ativo_id','nome','descricao','valor'])
                temp['ativo_id'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_aplicao_financeira = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/despesa_receita_descritivo.csv')
                df_aplicao_financeira = pd.concat([df_aplicao_financeira, temp], ignore_index=True)
                df_aplicao_financeira.to_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/despesa_receita_descritivo.csv',index=False)

            elif base == "Resultado Financeiro":
                temp = pd.DataFrame(columns=['ativo_id','nome','descricao','valor'])
                temp['ativo_id'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_contas_receber = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/resultado_fianceiro_descritivo.csv')
                df_contas_receber = pd.concat([df_contas_receber, temp], ignore_index=True)
                df_contas_receber.to_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/resultado_fianceiro_descritivo.csv',index=False)
            
            elif base == "Imposto de Renda e Contribuição Social sobre o Lucro":
                temp = pd.DataFrame(columns=['ativo_id','nome','descricao','valor'])
                temp['ativo_id'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_tributos_recuperar = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/imposto_descritivo.csv')
                df_tributos_recuperar = pd.concat([df_tributos_recuperar, temp], ignore_index=True)
                df_tributos_recuperar.to_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/imposto_descritivo.csv',index=False) 

            elif base == "Resultado Líquido de Operações Descontinuadas":
                temp = pd.DataFrame(columns=['ativo_id','nome','descricao','valor'])
                temp['ativo_id'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_resultado_descontinuado = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/resultado_descontinuado_descritivo.csv')
                df_resultado_descontinuado = pd.concat([df_resultado_descontinuado, temp], ignore_index=True)
                df_resultado_descontinuado.to_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/resultado_descontinuado_descritivo.csv',index=False) 

            elif base == "Lucro/Prejuízo Consolidado do Período":
                temp = pd.DataFrame(columns=['ativo_id','nome','descricao','valor'])
                temp['ativo_id'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_lucro_prejuizo_descritivo = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/resultado_descontinuado_descritivo.csv')
                df_lucro_prejuizo_descritivo = pd.concat([df_lucro_prejuizo_descritivo, temp], ignore_index=True)
                df_lucro_prejuizo_descritivo.to_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/resultado_descontinuado_descritivo.csv',index=False) 


        except Exception as e:
            logger.error("Erro na função inserir_ativo")
            logger.error(f"Erro ao processar item ({nome}, {descricao}, {valor}): {e}")
            break
    # con.close()

def separar_classes(procurar, resultado, ID):
    lista_prefixos = []
    lista_sufixos = []
    lista_valores = []

    caminhos_pos_procurar = []

    # Coletar todos os caminhos abaixo do "procurar"
    for chave in resultado:
        partes = chave.split(" | ")
        indices_procurar = [i for i, p in enumerate(partes) if p == procurar]
        idx = next((i for i in indices_procurar if i + 1 <= len(partes)), None)
        if idx is not None:
            caminho = partes[idx + 1:]
            caminhos_pos_procurar.append(caminho)

    # Determinar quais caminhos são folhas
    folhas = []
    for caminho in caminhos_pos_procurar:
        is_folha = not any(
            outro != caminho and outro[:len(caminho)] == caminho
            for outro in caminhos_pos_procurar
        )
        if is_folha:
            folhas.append(caminho)

    # Agora, coletar os valores somente dos nós folha
    for chave, valor in resultado.items():
        partes = chave.split(" | ")
        indices_procurar = [i for i, p in enumerate(partes) if p == procurar]
        idx = next((i for i in indices_procurar if i + 1 <= len(partes)), None)
        if idx is None:
            continue

        caminho = partes[idx + 1:]

        if caminho in folhas:
            if len(caminho) == 0:
                lista_prefixos.append(procurar)
                lista_sufixos.append(procurar)
            elif len(caminho) == 1:
                lista_prefixos.append(caminho[0])
                lista_sufixos.append(caminho[0])
            elif len(caminho) > 1:
                lista_prefixos.append(caminho[0])
                lista_sufixos.append(caminho[1])
            lista_valores.append(valor)
    # if lista_valores:
    #     inserir_ativo(procurar, lista_prefixos, lista_sufixos, lista_valores, ID)
    # else:
    #     logger.warning(f"Nenhum valor encontrado para base {procurar} no ID {ID}")

    inserir_ativo(procurar, lista_prefixos, lista_sufixos, lista_valores, ID)

def circulante(id=None, COMPANIA_ID=None, dados=None):
    valor_procurado = "Demonstração do Resultado"
    lista = [
        "Demonstração do Resultado | Receita de Venda de Bens e/ou Serviços",
        "Demonstração do Resultado | Custo dos Bens e/ou Serviços Vendidos",
        "Demonstração do Resultado | Resultado Bruto",
        "Demonstração do Resultado | Despesas/Receitas Operacionais",
        "Demonstração do Resultado | Resultado Antes do Resultado Financeiro e dos Tributos",
        "Demonstração do Resultado | Resultado Financeiro",
        "Demonstração do Resultado | Imposto de Renda e Contribuição Social sobre o Lucro",
        "Demonstração do Resultado | Resultado Líquido das Operações Continuadas",
        "Demonstração do Resultado | Resultado Líquido de Operações Descontinuadas",
        "Demonstração do Resultado | Lucro/Prejuízo do Período"
    ]

    try:
        resultado_atual = {}
        for chave, valor in dados.items():
            for procurado in lista:
                if procurado in chave:
                    resultado_atual[chave] = valor  # corrigido de 'va'

        for l in lista:
            partes = l.split(" | ")
            if len(partes) >= 2:
                procurar = partes[1]
                resultado = {}
                for chave, valor in resultado_atual.items():
                    if l in chave or f"{valor_procurado} | {procurar}" in chave:
                        resultado[chave] = valor

                separar_classes(procurar, resultado, id)  # corrigido dados_data → id

    except Exception as e:
        logger.error("Problema na função circulante")
        logger.error(e)

def salvar_valor_csv(caminho_csv, id, dados_data, chave):
    df = pd.read_csv(caminho_csv)
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados_data.get(chave, None)]  # Evita KeyError
    df = pd.concat([df, temp], ignore_index=True)
    df.to_csv(caminho_csv, index=False)

def inserir_resumo(dados, id):
    base_path = "/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/"

    salvar_valor_csv(f"{base_path}receita.csv", id, dados['data'], "Demonstração do Resultado | Receita de Venda de Bens e/ou Serviços")
    salvar_valor_csv(f"{base_path}custo.csv", id, dados['data'], "Demonstração do Resultado | Custo dos Bens e/ou Serviços Vendidos")
    salvar_valor_csv(f"{base_path}resultado_bruto.csv", id, dados['data'], "Demonstração do Resultado | Resultado Bruto")
    salvar_valor_csv(f"{base_path}despesa_receita.csv", id, dados['data'], "Demonstração do Resultado | Despesas/Receitas Operacionais")
    salvar_valor_csv(f"{base_path}resultado_antes.csv", id, dados['data'], "Demonstração do Resultado | Resultado Antes do Resultado Financeiro e dos Tributos")
    salvar_valor_csv(f"{base_path}resultado_fianceiro.csv", id, dados['data'], "Demonstração do Resultado | Resultado Financeiro")
    salvar_valor_csv(f"{base_path}imposto.csv", id, dados['data'], "Demonstração do Resultado | Imposto de Renda e Contribuição Social sobre o Lucro")
    salvar_valor_csv(f"{base_path}resultado_cinuados.csv", id, dados['data'], "Demonstração do Resultado | Resultado Líquido das Operações Continuadas")
    salvar_valor_csv(f"{base_path}resultado_descontinuado.csv", id, dados['data'], "Demonstração do Resultado | Resultado Líquido de Operações Descontinuadas")
    salvar_valor_csv(f"{base_path}lucro_prejuizo.csv", id, dados['data'], "Demonstração do Resultado | Lucro/Prejuízo Consolidado do Período")
```
