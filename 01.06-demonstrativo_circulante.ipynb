```python
def ativo_id():
    """
    Retorna o próximo ID para a tabela demonstrativo_resultado.ativo_total.
    O ID é gerado com base no maior valor já existente na tabela.
    Se a tabela estiver vazia, o ID começará em 1.
    """
    df = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/demonstrativo_resultado.csv')
    if df.empty:
        return 1
    else:
        return df['ativo_id'].max() + 1

def inserir_ativo(base, lista_prefixos, lista_sufixos, lista_valores, ID):
    # con = PostgresConnector()
    for nome, descricao, valor in zip(lista_prefixos, lista_sufixos, lista_valores):
        try:
            if base == "Despesas/Receitas Operacionais":
                temp = pd.DataFrame(columns=['ativo_id','nome','descricao','valor'])
                temp['ativo_id'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_aplicao_financeira = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/despesa_receita_descritivo.csv')
                df_aplicao_financeira = pd.concat([df_aplicao_financeira, temp], ignore_index=True)
                df_aplicao_financeira.to_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/despesa_receita_descritivo.csv',index=False)

            elif base == "Resultado Financeiros":
                temp = pd.DataFrame(columns=['ativo_id','nome','descricao','valor'])
                temp['ativo_id'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_contas_receber = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/resultado_fianceiro_descritivo.csv')
                df_contas_receber = pd.concat([df_contas_receber, temp], ignore_index=True)
                df_contas_receber.to_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/resultado_fianceiro_descritivo.csv',index=False)
            
            elif base == "Imposto de Renda e Contribuição Social sobre o Lucro":
                temp = pd.DataFrame(columns=['ativo_id','nome','descricao','valor'])
                temp['ativo_id'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_tributos_recuperar = pd.read_csv('//home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/imposto_descritivo.csv')
                df_tributos_recuperar = pd.concat([df_tributos_recuperar, temp], ignore_index=True)
                df_tributos_recuperar.to_csv('//home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/imposto_descritivo.csv',index=False) 

        except Exception as e:
            logger.error("Erro na função inserir_ativo")
            logger.error(f"Erro ao processar item ({nome}, {descricao}, {valor}): {e}")
            break
    # con.close()

def separar_classes(procurar, resultado, ID):
    lista_prefixos = []
    lista_sufixos = []
    lista_valores = []

    caminhos_pos_procurar = []

    # Coletar todos os caminhos abaixo do "procurar"
    for chave in resultado:
        partes = chave.split(" | ")
        indices_procurar = [i for i, p in enumerate(partes) if p == procurar]
        idx = next((i for i in indices_procurar if i + 1 <= len(partes)), None)
        if idx is not None:
            caminho = partes[idx + 1:]
            caminhos_pos_procurar.append(caminho)

    # Determinar quais caminhos são folhas
    folhas = []
    for caminho in caminhos_pos_procurar:
        is_folha = not any(
            outro != caminho and outro[:len(caminho)] == caminho
            for outro in caminhos_pos_procurar
        )
        if is_folha:
            folhas.append(caminho)

    # Agora, coletar os valores somente dos nós folha
    for chave, valor in resultado.items():
        partes = chave.split(" | ")
        indices_procurar = [i for i, p in enumerate(partes) if p == procurar]
        idx = next((i for i in indices_procurar if i + 1 <= len(partes)), None)
        if idx is None:
            continue

        caminho = partes[idx + 1:]

        if caminho in folhas:
            if len(caminho) == 0:
                lista_prefixos.append(procurar)
                lista_sufixos.append(procurar)
            elif len(caminho) == 1:
                lista_prefixos.append(caminho[0])
                lista_sufixos.append(caminho[0])
            elif len(caminho) > 1:
                lista_prefixos.append(caminho[0])
                lista_sufixos.append(caminho[1])
            lista_valores.append(valor)

    inserir_ativo(procurar, lista_prefixos, lista_sufixos, lista_valores, ID)

def circulante( id=None, COMPANIA_ID=None, dados=None):

    valor_procurado = "Demonstração do Resultado"
    lista = ["Demonstração do Resultado | Receita de Venda de Bens e/ou Serviços",
                 "Demonstração do Resultado | Custo dos Bens e/ou Serviços Vendidos",
                 "Demonstração do Resultado | Resultado Bruto",
                 "Demonstração do Resultado | Despesas/Receitas Operacionais"
                 "Demonstração do Resultado | Resultado Antes do Resultado Financeiro e dos Tributos"
                 "Demonstração do Resultado | Resultado Financeiro"
                 "Demonstração do Resultado | Imposto de Renda e Contribuição Social sobre o Lucro",
                 "Demonstração do Resultado | Resultado Líquido das Operações Continuadas",
                 "Demonstração do Resultado | Resultado Líquido de Operações Descontinuadas",
                 "Demonstração do Resultado | Lucro/Prejuízo do Período"
                 ]


    resultado_atual = {}

    for chave, valor in dados.items():
        if valor_procurado in chave:
            resultado_atual[chave] = valor

    for l in lista:
        valor_procurado = l
        partes = valor_procurado.split(" | ")
        if len(partes) >= 3:
            procurar = partes[2]

        resultado = {}

        for chave, valor in resultado_atual.items():
            if l in chave:
                resultado[chave] = valor

        separar_classes(procurar, resultado, id)

def inserir_resumo(dados,id):   
    obr_sociareceitais_trabalhista = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/receita.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Demonstração do Resultado | Receita de Venda de Bens e/ou Serviços"]]
    receita = pd.concat([receita,temp])
    receita.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/receita.csv",index=False)

    custo = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/custo.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Demonstração do Resultado | Custo dos Bens e/ou Serviços Vendidos"]]
    custo = pd.concat([custo,temp])
    custo.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/custo.csv",index=False)

    resultado_bruto = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/resultado_bruto.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Demonstração do Resultado | Resultado Bruto"]]
    resultado_bruto = pd.concat([resultado_bruto,temp])
    resultado_bruto.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/resultado_bruto.csv",index=False)

    despesa_receita = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/despesa_receita.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Demonstração do Resultado | Despesas/Receitas Operacionais"]]
    despesa_receita = pd.concat([despesa_receita,temp])
    despesa_receita.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/despesa_receita.csv",index=False)

    resultado_antes = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/resultado_antes.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Demonstração do Resultado | Resultado Antes do Resultado Financeiro e dos Tributos"]]
    resultado_antes = pd.concat([resultado_antes,temp])
    resultado_antes.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/resultado_antes.csv",index=False)

    resultado_fianceiro = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/resultado_fianceiro.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Demonstração do Resultado | Resultado Financeiro"]]
    resultado_fianceiro = pd.concat([resultado_fianceiro,temp])
    resultado_fianceiro.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/resultado_fianceiro.csv",index=False)

    imposto = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/imposto.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Demonstração do Resultado | Imposto de Renda e Contribuição Social sobre o Lucro"]]
    imposto = pd.concat([imposto,temp])
    imposto.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/imposto.csv",index=False)    
    
    resultado_cinuados = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/resultado_cinuados.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Demonstração do Resultado | Imposto de Renda e Contribuição Social sobre o Lucro"]]
    resultado_cinuados = pd.concat([resultado_cinuados,temp])
    resultado_cinuados.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/resultado_cinuados.csv",index=False)  

    resultado_descontinuado = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/resultado_descontinuado.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Demonstração do Resultado | Imposto de Renda e Contribuição Social sobre o Lucro"]]
    resultado_descontinuado = pd.concat([resultado_descontinuado,temp])
    resultado_descontinuado.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/resultado_descontinuado.csv",index=False)  

    lucro_prejuizo = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/lucro_prejuizo.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Demonstração do Resultado | Lucro/Prejuízo do Período"]]
    lucro_prejuizo = pd.concat([lucro_prejuizo,temp])
    lucro_prejuizo.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/lucro_prejuizo.csv",index=False)  
```
