```python
import pandas as pd
```

# Funções

```python
def juntar_dataframe(df1,df2,nome):
    df2 = df2.groupby(['codigo'])['valor'].sum().reset_index()
    df = pd.merge(
    df1,
    df2[['codigo', 'valor']],
    on='codigo',
    how='left'
    )
    try:
        df.drop(columns=['start_date','end_date','retrieval_date','publish_date'], inplace=True)
    except:
        pass
    df = df.rename(columns={'valor': nome})
    return df
```

# Ativo Circulante

```python
ativo_total = pd.read_csv("data/dataframe/ativo_circulante/ativo_total.csv")
aplicacao_financeira = pd.read_csv("data/dataframe/ativo_circulante/aplicao_financeira.csv")
ativos_biologicos = pd.read_csv("data/dataframe/ativo_circulante/ativos_biologicos.csv")
caixa = pd.read_csv("data/dataframe/ativo_circulante/caixa.csv")
contas_receber = pd.read_csv("data/dataframe/ativo_circulante/contas_receber.csv")
despesas_antecipadas = pd.read_csv("data/dataframe/ativo_circulante/despesas_antecipadas.csv")
estoques = pd.read_csv("data/dataframe/ativo_circulante/estoque.csv")
outros_ativos_circulantes = pd.read_csv("data/dataframe/ativo_circulante/outros_ativos_circulantes.csv")
tributos_recuperar = pd.read_csv("data/dataframe/ativo_circulante/tributos_recuperar.csv")
```

```python
ativo_circulante = juntar_dataframe(ativo_total,aplicacao_financeira,'aplicacao_financeira')
ativo_circulante = juntar_dataframe(ativo_circulante,ativos_biologicos,'ativo_biologico')
ativo_circulante = juntar_dataframe(ativo_circulante,caixa,'caixa')
ativo_circulante = juntar_dataframe(ativo_circulante,contas_receber,'contas_receber')
ativo_circulante = juntar_dataframe(ativo_circulante,despesas_antecipadas,'despesas_antecipadas')
ativo_circulante = juntar_dataframe(ativo_circulante,estoques,'estoques')
ativo_circulante = juntar_dataframe(ativo_circulante,outros_ativos_circulantes,'outros_ativos_circulantes')
ativo_circulante = juntar_dataframe(ativo_circulante,tributos_recuperar,'tributos_recuperar')
```

```python
colunas_ativo_circulante = [
    'aplicacao_financeira',
    'ativo_biologico',
    'caixa',
    'contas_receber',
    'despesas_antecipadas',
    'estoques',
    'outros_ativos_circulantes',
    'tributos_recuperar'
]
ativo_circulante['total'] = ativo_circulante[colunas_ativo_circulante].sum(axis=1)
```
