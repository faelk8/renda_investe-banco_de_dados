```python
def ativo_id():
    """
    Retorna o próximo ID para a tabela ativo_nao_circulante.ativo_total.
    O ID é gerado com base no maior valor já existente na tabela.
    Se a tabela estiver vazia, o ID começará em 1.
    """
    df = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/ativo_nao_circulante/ativo_nao_circulante.csv')
    if df.empty:
        return 1
    else:
        return df['ativo_id'].max() + 1

def inserir_ativo(base, lista_prefixos, lista_sufixos, lista_valores, ID):
    # con = PostgresConnector()
    for nome, descricao, valor in zip(lista_prefixos, lista_sufixos, lista_valores):
        try:
            if base == "Ativo Realizável a Longo Prazo":
                temp = pd.DataFrame(columns=['ativo_id','nome','descricao','valor'])
                temp['ativo_id'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_realizavel_longo_prazo_descritivo = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/ativo_nao_circulante/realizavel_longo_prazo_descritivo.csv')
                df_realizavel_longo_prazo_descritivo = pd.concat([df_realizavel_longo_prazo_descritivo, temp], ignore_index=True)
                df_realizavel_longo_prazo_descritivo.to_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/ativo_nao_circulante/realizavel_longo_prazo_descritivo.csv',index=False)
                # tabela = "ativo_nao_circulante.aplicao_financeira"
                # key = "aplicao_financeira_id"

            elif base == "Investimentos":
                temp = pd.DataFrame(columns=['ativo_id','nome','descricao','valor'])
                temp['ativo_id'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_investimento_descritivo = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/ativo_nao_circulante/investimento_descritivo.csv')
                df_investimento_descritivo = pd.concat([df_investimento_descritivo, temp], ignore_index=True)
                df_investimento_descritivo.to_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/ativo_nao_circulante/investimento_descritivo.csv',index=False)
                # tabela = "ativo_nao_circulante.contas_receber"
                # key = "contas_receber_id"
           
            elif base == "Imobilizado":
                temp = pd.DataFrame(columns=['ativo_id','nome','descricao','valor'])
                temp['ativo_id'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_imobilizado_descritivo = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/ativo_nao_circulante/imobilizado_descritivo.csv')
                df_imobilizado_descritivo = pd.concat([df_imobilizado_descritivo, temp], ignore_index=True)
                df_imobilizado_descritivo.to_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/ativo_nao_circulante/imobilizado_descritivo.csv',index=False)
                # tabela = "ativo_nao_circulante.imobilizado_recuperar"
                # key = "imobilizado_recuperar_id"
            
            elif base == "Intangível":
                temp = pd.DataFrame(columns=['ativo_id','nome','descricao','valor'])
                temp['ativo_id'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_intangivel_descritivo = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/ativo_nao_circulante/intangivel_descritivo.csv')
                df_intangivel_descritivo = pd.concat([df_intangivel_descritivo, temp], ignore_index=True)
                df_intangivel_descritivo.to_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/ativo_nao_circulante/intangivel_descritivo.csv',index=False)                # tabela = "ativo_nao_circulante.intangivel"
                # key = "intangivel_id"

        except Exception as e:
            logger.error("Erro na função inserir_ativo")
            logger.error(f"Erro ao processar item ({nome}, {descricao}, {valor}): {e}")
            break
    # con.close()

def separar_classes(procurar, resultado, ID):
    lista_prefixos = []
    lista_sufixos = []
    lista_valores = []

    caminhos_pos_procurar = []

    # Coletar todos os caminhos abaixo do "procurar"
    for chave in resultado:
        partes = chave.split(" | ")
        indices_procurar = [i for i, p in enumerate(partes) if p == procurar]
        idx = next((i for i in indices_procurar if i + 1 <= len(partes)), None)
        if idx is not None:
            caminho = partes[idx + 1:]
            caminhos_pos_procurar.append(caminho)

    # Determinar quais caminhos são folhas
    folhas = []
    for caminho in caminhos_pos_procurar:
        is_folha = not any(
            outro != caminho and outro[:len(caminho)] == caminho
            for outro in caminhos_pos_procurar
        )
        if is_folha:
            folhas.append(caminho)

    # Agora, coletar os valores somente dos nós folha
    for chave, valor in resultado.items():
        partes = chave.split(" | ")
        indices_procurar = [i for i, p in enumerate(partes) if p == procurar]
        idx = next((i for i in indices_procurar if i + 1 <= len(partes)), None)
        if idx is None:
            continue

        caminho = partes[idx + 1:]

        if caminho in folhas:
            if len(caminho) == 0:
                lista_prefixos.append(procurar)
                lista_sufixos.append(procurar)
            elif len(caminho) == 1:
                lista_prefixos.append(caminho[0])
                lista_sufixos.append(caminho[0])
            elif len(caminho) > 1:
                lista_prefixos.append(caminho[0])
                lista_sufixos.append(caminho[1])
            lista_valores.append(valor)

    inserir_ativo(procurar, lista_prefixos, lista_sufixos, lista_valores, ID)

def circulante( id=None, COMPANIA_ID=None, dados=None):

    valor_procurado = "Ativo Total | Ativo Não Circulante"
    lista  = ["Ativo Total | Ativo Não Circulante | Ativo Realizável a Longo Prazo",
                "Ativo Total | Ativo Não Circulante | Investimentos",
                "Ativo Total | Ativo Não Circulante | Imobilizado",
                "Ativo Total | Ativo Não Circulante | Intangível",
                ]


    resultado_atual = {}

    for chave, valor in dados.items():
        if valor_procurado in chave:
            resultado_atual[chave] = valor

    for l in lista:
        valor_procurado = l
        partes = valor_procurado.split(" | ")
        if len(partes) >= 3:
            procurar = partes[2]

        resultado = {}

        for chave, valor in resultado_atual.items():
            if l in chave:
                resultado[chave] = valor

        separar_classes(procurar, resultado, id)

def inserir_resumo(dados,id):   
    realizavel_longo_prazo = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/ativo_nao_circulante/realizavel_longo_prazo.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Ativo Total | Ativo Não Circulante | Ativo Realizável a Longo Prazo"]]
    realizavel_longo_prazo = pd.concat([realizavel_longo_prazo,temp])
    realizavel_longo_prazo.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/ativo_nao_circulante/realizavel_longo_prazo.csv",index=False)

    investimento = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/ativo_nao_circulante/investimento.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Ativo Total | Ativo Não Circulante | Investimentos"]]
    investimento = pd.concat([investimento,temp])
    investimento.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/ativo_nao_circulante/investimento.csv",index=False)

    imobilizado = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/ativo_nao_circulante/imobilizado.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Ativo Total | Ativo Não Circulante | Imobilizado"]]
    imobilizado = pd.concat([imobilizado,temp])
    imobilizado.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/ativo_nao_circulante/imobilizado.csv",index=False)

    intangivel = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/ativo_nao_circulante/intangivel.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Ativo Total | Ativo Não Circulante | Intangível"]]
    intangivel = pd.concat([intangivel,temp])
    intangivel.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/ativo_nao_circulante/intangivel.csv",index=False)

           
```
