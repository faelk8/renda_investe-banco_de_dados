```python
def ativo_id():
    """
    Retorna o próximo ID para a tabela ativo_circulante.ativo_total.
    O ID é gerado com base no maior valor já existente na tabela.
    Se a tabela estiver vazia, o ID começará em 1.
    """
    df = pd.read_csv('data/dataframe/passivo_nao_circulante/passivo_total.csv')
    if df.empty:
        return 1
    else:
        return df['codigo'].max() + 1

def inserir_ativo(base, lista_prefixos, lista_sufixos, lista_valores, ID):
    # con = PostgresConnector()
    for nome, descricao, valor in zip(lista_prefixos, lista_sufixos, lista_valores):
        try:
            if  base == "Empréstimos e Financiamentos":
                temp = pd.DataFrame(columns=['codigo','nome','descricao','valor'])
                temp['codigo'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_emprestimos_financiamentos = pd.read_csv('data/dataframe/passivo_nao_circulante/emprestimos_financiamentos.csv')
                df_emprestimos_financiamentos = pd.concat([df_emprestimos_financiamentos, temp], ignore_index=True)
                df_emprestimos_financiamentos.to_csv('data/dataframe/passivo_nao_circulante/emprestimos_financiamentos.csv',index=False)
                # tabela = "passivo_nao_circulante.emprestimos_financiamentos"
                # key = "emprestimos_financiamentos_id"

            elif base == "Lucros e Receitas a Apropriar":
                temp = pd.DataFrame(columns=['codigo','nome','descricao','valor'])
                temp['codigo'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_lucro_receita = pd.read_csv('data/dataframe/passivo_nao_circulante/lucro_receita.csv')
                df_lucro_receita = pd.concat([df_lucro_receita, temp], ignore_index=True)
                df_lucro_receita.to_csv('data/dataframe/passivo_nao_circulante/lucro_receita.csv',index=False)
                # tabela = "passivo_nao_circulante.lucro_receita"
                # key = "lucro_receita_id"
            
            elif base == "Outras Obrigações":
                temp = pd.DataFrame(columns=['codigo','nome','descricao','valor'])
                temp['codigo'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_outras_obrigacoes = pd.read_csv('data/dataframe/passivo_nao_circulante/outras_obrigacoes.csv')
                df_outras_obrigacoes = pd.concat([df_outras_obrigacoes, temp], ignore_index=True)
                df_outras_obrigacoes.to_csv('data/dataframe/passivo_nao_circulante/outras_obrigacoes.csv',index=False)
                # tabela = "passivo_nao_circulante.outras_obrigacoes"
                # key = "outras_obrigacoes_id"

            elif base == "Passivos sobre Ativos Não-Correntes a Venda e Descontinuados":
                temp = pd.DataFrame(columns=['codigo','nome','descricao','valor'])
                temp['codigo'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_passivos = pd.read_csv('data/dataframe/passivo_nao_circulante/passivos.csv')
                df_passivos = pd.concat([df_passivos, temp], ignore_index=True)
                df_passivos.to_csv('data/dataframe/passivo_nao_circulante/passivos.csv',index=False)
                # tabela = "passivo_nao_circulante.passivos"
                # key = "passivos_id"
            
            elif base == "Provisões":
                temp = pd.DataFrame(columns=['codigo','nome','descricao','valor'])
                temp['codigo'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_provisoes = pd.read_csv('data/dataframe/passivo_nao_circulante/provisoes.csv')
                df_provisoes = pd.concat([df_provisoes, temp], ignore_index=True)
                df_provisoes.to_csv('data/dataframe/passivo_nao_circulante/provisoes.csv',index=False)
                # tabela = "passivo_nao_circulante.provisoes"
                # key = "provisoes_id"
            
            elif base == "Tributos Diferidos":
                temp = pd.DataFrame(columns=['codigo','nome','descricao','valor'])
                temp['codigo'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_tibrutos_diferidos = pd.read_csv('data/dataframe/passivo_nao_circulante/tibrutos_diferidos.csv')
                df_tibrutos_diferidos = pd.concat([df_tibrutos_diferidos, temp], ignore_index=True)
                df_tibrutos_diferidos.to_csv('data/dataframe/passivo_nao_circulante/tibrutos_diferidos.csv',index=False)                # tabela = "passivo_nao_circulante.passivos"
                # key = "passivos_id"

        except Exception as e:
            logger.error("Erro na função inserir_ativo")
            logger.error(f"Erro ao processar item ({nome}, {descricao}, {valor}): {e}")
            break
    # con.close()

def separar_classes(procurar, resultado, ID):
    lista_prefixos = []
    lista_sufixos = []
    lista_valores = []

    caminhos_pos_procurar = []

    # Coletar todos os caminhos abaixo do "procurar"
    for chave in resultado:
        partes = chave.split(" | ")
        indices_procurar = [i for i, p in enumerate(partes) if p == procurar]
        idx = next((i for i in indices_procurar if i + 1 <= len(partes)), None)
        if idx is not None:
            caminho = partes[idx + 1:]
            caminhos_pos_procurar.append(caminho)

    # Determinar quais caminhos são folhas
    folhas = []
    for caminho in caminhos_pos_procurar:
        is_folha = not any(
            outro != caminho and outro[:len(caminho)] == caminho
            for outro in caminhos_pos_procurar
        )
        if is_folha:
            folhas.append(caminho)

    # Agora, coletar os valores somente dos nós folha
    for chave, valor in resultado.items():
        partes = chave.split(" | ")
        indices_procurar = [i for i, p in enumerate(partes) if p == procurar]
        idx = next((i for i in indices_procurar if i + 1 <= len(partes)), None)
        if idx is None:
            continue

        caminho = partes[idx + 1:]

        if caminho in folhas:
            if len(caminho) == 0:
                lista_prefixos.append(procurar)
                lista_sufixos.append(procurar)
            elif len(caminho) == 1:
                lista_prefixos.append(caminho[0])
                lista_sufixos.append(caminho[0])
            elif len(caminho) > 1:
                lista_prefixos.append(caminho[0])
                lista_sufixos.append(caminho[1])
            lista_valores.append(valor)

    inserir_ativo(procurar, lista_prefixos, lista_sufixos, lista_valores, ID)
    # print(procurar, lista_prefixos, lista_sufixos, lista_valores, ID)

def circulante( id=None, COMPANIA_ID=None, dados=None):

    valor_procurado = "Passivo Total | Passivo Não Circulante"
    lista = ["Passivo Total | Passivo Não Circulante | Empréstimos e Financiamentos",
                 "Passivo Total | Passivo Não Circulante | Outras Obrigações",
                 "Passivo Total | Passivo Não Circulante | Tributos Diferidos",
                 "Passivo Total | Passivo Não Circulante | Provisões",
                 "Passivo Total | Passivo Não Circulante | Passivos sobre Ativos Não-Correntes a Venda e Descontinuados",
                 "Passivo Total | Passivo Não Circulante | Lucros e Receitas a Apropriar"
                ]

    resultado_atual = {}

    for chave, valor in dados.items():
        if valor_procurado in chave:
            resultado_atual[chave] = valor

    for l in lista:
        valor_procurado = l
        partes = valor_procurado.split(" | ")
        if len(partes) >= 3:
            procurar = partes[2]

        resultado = {}

        for chave, valor in resultado_atual.items():
            if l in chave:
                resultado[chave] = valor
        separar_classes(procurar, resultado, id)
```
