```python
import pandas as pd
```

# Funções

```python
def juntar_dataframe(df1,df2,nome2,nome1=None,):
    try :
        df1 = df1.rename(columns={'valor': nome1})
    except:
        pass
    df = pd.merge(df1, df2, on='ativo_id', how='left' )
    try:
        df.drop(columns=['start_date','end_date','retrieval_date','publish_date'], inplace=True)
    except:
        pass
    df = df.rename(columns={'valor': nome2})
    return df
```
