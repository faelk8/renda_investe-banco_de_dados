# Funções

```python
def juntar_dataframe(df1,df2,nome2,nome1=None,):
    try :
        df1 = df1.rename(columns={'valor': nome1})
    except:
        pass
    df = pd.merge(df1, df2, on='ativo_id', how='left' )
    try:
        df.drop(columns=['start_date','end_date','retrieval_date','publish_date'], inplace=True)
    except:
        pass
    df = df.rename(columns={'valor': nome2})
    return df
```

```python
def janela_tempo(dados, coluna):
    somas = []
    janela = 4

    # Loop pelas janelas
    for i in range(len(dados) - janela + 1):
        soma = dados[coluna].iloc[i:i+janela].sum()
        somas.append(soma)

    # Preencher com zeros no final
    somas += [0] * (len(dados) - len(somas))

    return somas
```

```python
ebitda = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/demonstrativo_resultado/despesa_receita_descritivo.csv")

eb1 = ebitda[(ebitda['descricao']=='Depreciações e Amortizações')]
eb2 = ebitda[(ebitda['descricao']=='Depreciações e Amortizações de Direito de Uso')]
ebitda = pd.concat([eb1,eb2])

ebitda = ebitda.groupby('ativo_id')['valor'].sum().reset_index()

ebitda.rename(columns={'valor':'ebitda'}, inplace=True)
```
