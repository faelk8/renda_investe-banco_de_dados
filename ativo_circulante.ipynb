```python
def ativo_id():
    """
    Retorna o próximo ID para a tabela ativo_circulante.ativo_total.
    O ID é gerado com base no maior valor já existente na tabela.
    Se a tabela estiver vazia, o ID começará em 1.
    """
    df = pd.read_csv('data/dataframe/ativo_circulante/ativo_total.csv')
    if df.empty:
        return 1
    else:
        return df['codigo'].max() + 1

def inserir_ativo(base, lista_prefixos, lista_sufixos, lista_valores, ID):
    # con = PostgresConnector()
    for nome, descricao, valor in zip(lista_prefixos, lista_sufixos, lista_valores):
        try:
            if base == "Caixa e Equivalentes de Caixa":
                temp = pd.DataFrame(columns=['codigo','nome','descricao','valor'])
                temp['codigo'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_caixa = pd.read_csv('data/dataframe/ativo_circulante/caixa.csv')
                df_caixa = pd.concat([df_caixa, temp], ignore_index=True)
                df_caixa.to_csv('data/dataframe/ativo_circulante/caixa.csv',index=False)
                # tabela = "ativo_circulante.caixa"
                # key = "caixa_id"

            elif base == "Aplicações Financeiras":
                temp = pd.DataFrame(columns=['codigo','nome','descricao','valor'])
                temp['codigo'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_aplicao_financeira = pd.read_csv('data/dataframe/ativo_circulante/aplicao_financeira.csv')
                df_aplicao_financeira = pd.concat([df_aplicao_financeira, temp], ignore_index=True)
                df_aplicao_financeira.to_csv('data/dataframe/ativo_circulante/aplicao_financeira.csv',index=False)
                # tabela = "ativo_circulante.aplicao_financeira"
                # key = "aplicao_financeira_id"

            elif base == "Contas a Receber":
                temp = pd.DataFrame(columns=['codigo','nome','descricao','valor'])
                temp['codigo'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_contas_receber = pd.read_csv('data/dataframe/ativo_circulante/contas_receber.csv')
                df_contas_receber = pd.concat([df_contas_receber, temp], ignore_index=True)
                df_contas_receber.to_csv('data/dataframe/ativo_circulante/contas_receber.csv',index=False)
                # tabela = "ativo_circulante.contas_receber"
                # key = "contas_receber_id"
            
            elif base == "Estoques":
                temp = pd.DataFrame(columns=['codigo','nome','descricao','valor'])
                temp['codigo'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_estoque = pd.read_csv('data/dataframe/ativo_circulante/estoque.csv')
                df_estoque = pd.concat([df_estoque, temp], ignore_index=True)
                df_estoque.to_csv('data/dataframe/ativo_circulante/estoque.csv',index=False)
                # tabela = "ativo_circulante.estoque"
                # key = "estoque_id"
            
            elif base == "Ativos Biológicos":
                temp = pd.DataFrame(columns=['codigo','nome','descricao','valor'])
                temp['codigo'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_ativos_biologicos = pd.read_csv('data/dataframe/ativo_circulante/ativos_biologicos.csv')
                df_ativos_biologicos = pd.concat([df_ativos_biologicos, temp], ignore_index=True)
                df_ativos_biologicos.to_csv('data/dataframe/ativo_circulante/ativos_biologicos.csv',index=False)
                # tabela = "ativo_circulante.ativos_biologicos"
                # key = "ativos_biologicos_id"
            
            elif base == "Tributos a Recuperar":
                temp = pd.DataFrame(columns=['codigo','nome','descricao','valor'])
                temp['codigo'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_tributos_recuperar = pd.read_csv('data/dataframe/ativo_circulante/tributos_recuperar.csv')
                df_tributos_recuperar = pd.concat([df_tributos_recuperar, temp], ignore_index=True)
                df_tributos_recuperar.to_csv('data/dataframe/ativo_circulante/tributos_recuperar.csv',index=False)
                # tabela = "ativo_circulante.tributos_recuperar"
                # key = "tributos_recuperar_id"
            
            elif base == "Despesas Antecipadas":
                temp = pd.DataFrame(columns=['codigo','nome','descricao','valor'])
                temp['codigo'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_despesas_antecipadas = pd.read_csv('data/dataframe/ativo_circulante/despesas_antecipadas.csv')
                df_despesas_antecipadas = pd.concat([df_despesas_antecipadas, temp], ignore_index=True)
                df_despesas_antecipadas.to_csv('data/dataframe/ativo_circulante/despesas_antecipadas.csv',index=False)
                # tabela = "ativo_circulante.despesas_antecipadas"
                # key = "despesas_antecipadas_id"
            
            elif base == "Outros Ativos Circulantes":
                temp = pd.DataFrame(columns=['codigo','nome','descricao','valor'])
                temp['codigo'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_outros_ativos_circulantes = pd.read_csv('data/dataframe/ativo_circulante/outros_ativos_circulantes.csv')
                df_outros_ativos_circulantes = pd.concat([df_outros_ativos_circulantes, temp], ignore_index=True)
                df_outros_ativos_circulantes.to_csv('data/dataframe/ativo_circulante/outros_ativos_circulantes.csv',index=False)                # tabela = "ativo_circulante.outros_ativos_circulantes"
                # key = "outros_ativos_circulantes_id"

        except Exception as e:
            logger.error("Erro na função inserir_ativo")
            logger.error(f"Erro ao processar item ({nome}, {descricao}, {valor}): {e}")
            break
    # con.close()

def separar_classes(procurar, resultado, ID):
    lista_prefixos = []
    lista_sufixos = []
    lista_valores = []

    caminhos_pos_procurar = []

    # Coletar todos os caminhos abaixo do "procurar"
    for chave in resultado:
        partes = chave.split(" | ")
        indices_procurar = [i for i, p in enumerate(partes) if p == procurar]
        idx = next((i for i in indices_procurar if i + 1 <= len(partes)), None)
        if idx is not None:
            caminho = partes[idx + 1:]
            caminhos_pos_procurar.append(caminho)

    # Determinar quais caminhos são folhas
    folhas = []
    for caminho in caminhos_pos_procurar:
        is_folha = not any(
            outro != caminho and outro[:len(caminho)] == caminho
            for outro in caminhos_pos_procurar
        )
        if is_folha:
            folhas.append(caminho)

    # Agora, coletar os valores somente dos nós folha
    for chave, valor in resultado.items():
        partes = chave.split(" | ")
        indices_procurar = [i for i, p in enumerate(partes) if p == procurar]
        idx = next((i for i in indices_procurar if i + 1 <= len(partes)), None)
        if idx is None:
            continue

        caminho = partes[idx + 1:]

        if caminho in folhas:
            if len(caminho) == 0:
                lista_prefixos.append(procurar)
                lista_sufixos.append(procurar)
            elif len(caminho) == 1:
                lista_prefixos.append(caminho[0])
                lista_sufixos.append(caminho[0])
            elif len(caminho) > 1:
                lista_prefixos.append(caminho[0])
                lista_sufixos.append(caminho[1])
            lista_valores.append(valor)

    inserir_ativo(procurar, lista_prefixos, lista_sufixos, lista_valores, ID)

def circulante( id=None, COMPANIA_ID=None, dados=None):

    valor_procurado = "Ativo Total | Ativo Circulante"
    lista  = ["Ativo Total | Ativo Circulante | Caixa e Equivalentes de Caixa",
                "Ativo Total | Ativo Circulante | Aplicações Financeiras",
                "Ativo Total | Ativo Circulante | Contas a Receber",
                "Ativo Total | Ativo Circulante | Estoques",
                "Ativo Total | Ativo Circulante | Ativos Biológicos",
                "Ativo Total | Ativo Circulante | Tributos a Recuperar",
                "Ativo Total | Ativo Circulante | Despesas Antecipadas",
                "Ativo Total | Ativo Circulante | Outros Ativos Circulantes"
                ]


    resultado_atual = {}

    for chave, valor in dados.items():
        if valor_procurado in chave:
            resultado_atual[chave] = valor

    for l in lista:
        valor_procurado = l
        partes = valor_procurado.split(" | ")
        if len(partes) >= 3:
            procurar = partes[2]

        resultado = {}

        for chave, valor in resultado_atual.items():
            if l in chave:
                resultado[chave] = valor

        separar_classes(procurar, resultado, id)
```
