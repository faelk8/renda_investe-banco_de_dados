```python
def ativo_id():
    """
    Retorna o próximo ID para a tabela patrimonio_liquido_consolidados.ativo_total.
    O ID é gerado com base no maior valor já existente na tabela.
    Se a tabela estiver vazia, o ID começará em 1.
    """
    df = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio/patrimonio_liquido_consolidados.csv')
    if df.empty:
        return 1
    else:
        return df['ativo_id'].max() + 1

def inserir_ativo(base, lista_prefixos, lista_sufixos, lista_valores, ID):
    # con = PostgresConnector()
    for nome, descricao, valor in zip(lista_prefixos, lista_sufixos, lista_valores):
        try:
            if base == "Capital Social Realizado":
                temp = pd.DataFrame(columns=['ativo_id','nome','descricao','valor'])
                temp['ativo_id'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_capital_social_descritivo = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio/capital_social_descritivo.csv')
                df_capital_social_descritivo = pd.concat([df_capital_social_descritivo, temp], ignore_index=True)
                df_capital_social_descritivo.to_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio/capital_social_descritivo.csv',index=False)

            elif base == "Reservas de Capital":
                temp = pd.DataFrame(columns=['ativo_id','nome','descricao','valor'])
                temp['ativo_id'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_reserva_descritivo = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio/reserva_descritivo.csv')
                df_reserva_descritivo = pd.concat([df_reserva_descritivo, temp], ignore_index=True)
                df_reserva_descritivo.to_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio/reserva_descritivo.csv',index=False)
           
            elif "Reservas de Lucros":
                temp = pd.DataFrame(columns=['ativo_id','nome','descricao','valor'])
                temp['ativo_id'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_lucro_prejuizo_descritivo = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio/reserva_lucro_descritivo.csv')
                df_lucro_prejuizo_descritivo = pd.concat([df_lucro_prejuizo_descritivo, temp], ignore_index=True)
                df_lucro_prejuizo_descritivo.to_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio/reserva_lucro_descritivo.csv',index=False)

        except Exception as e:
            logger.error("Erro na função inserir_ativo")
            logger.error(f"Erro ao processar item ({nome}, {descricao}, {valor}): {e}")
            break
    # con.close()

def separar_classes(procurar, resultado, ID):
    lista_prefixos = []
    lista_sufixos = []
    lista_valores = []

    caminhos_pos_procurar = []

    # Coletar todos os caminhos abaixo do "procurar"
    for chave in resultado:
        partes = chave.split(" | ")
        indices_procurar = [i for i, p in enumerate(partes) if p == procurar]
        idx = next((i for i in indices_procurar if i + 1 <= len(partes)), None)
        if idx is not None:
            caminho = partes[idx + 1:]
            caminhos_pos_procurar.append(caminho)

    # Determinar quais caminhos são folhas
    folhas = []
    for caminho in caminhos_pos_procurar:
        is_folha = not any(
            outro != caminho and outro[:len(caminho)] == caminho
            for outro in caminhos_pos_procurar
        )
        if is_folha:
            folhas.append(caminho)

    # Agora, coletar os valores somente dos nós folha
    for chave, valor in resultado.items():
        partes = chave.split(" | ")
        indices_procurar = [i for i, p in enumerate(partes) if p == procurar]
        idx = next((i for i in indices_procurar if i + 1 <= len(partes)), None)
        if idx is None:
            continue

        caminho = partes[idx + 1:]

        if caminho in folhas:
            if len(caminho) == 0:
                lista_prefixos.append(procurar)
                lista_sufixos.append(procurar)
            elif len(caminho) == 1:
                lista_prefixos.append(caminho[0])
                lista_sufixos.append(caminho[0])
            elif len(caminho) > 1:
                lista_prefixos.append(caminho[0])
                lista_sufixos.append(caminho[1])
            lista_valores.append(valor)

    inserir_ativo(procurar, lista_prefixos, lista_sufixos, lista_valores, ID)

def circulante( id=None, COMPANIA_ID=None, dados=None):

    valor_procurado = "Passivo Total | Patrimônio Líquido Consolidado"
    lista = ["Passivo Total | Patrimônio Líquido Consolidado | Capital Social Realizado",
                 "Passivo Total | Patrimônio Líquido Consolidado | Reservas de Capital",
                 "Passivo Total | Patrimônio Líquido Consolidado | Reservas de Reavaliação",
                 "Passivo Total | Patrimônio Líquido Consolidado | Reservas de Lucros",
                 "Passivo Total | Patrimônio Líquido Consolidado | Lucros/Prejuízos Acumulados",
                 "Passivo Total | Patrimônio Líquido Consolidado | Ajustes de Avaliação Patrimonial",
                 "Passivo Total | Patrimônio Líquido Consolidado | Ajustes Acumulados de Conversão",
                 "Passivo Total | Patrimônio Líquido Consolidado | Outros Resultados Abrangentes",
                 "Passivo Total | Patrimônio Líquido Consolidado | Participação dos Acionistas Não Controladores"
                ]

    resultado_atual = {}

    for chave, valor in dados.items():
        if valor_procurado in chave:
            resultado_atual[chave] = valor

    for l in lista:
        valor_procurado = l
        partes = valor_procurado.split(" | ")
        if len(partes) >= 3:
            procurar = partes[2]

        resultado = {}

        for chave, valor in resultado_atual.items():
            if l in chave:
                resultado[chave] = valor

        separar_classes(procurar, resultado, id)

def inserir_resumo(dados,id):   
    capital_social = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio/capital_social.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Passivo Total | Patrimônio Líquido Consolidado | Capital Social Realizado"]]
    capital_social = pd.concat([capital_social,temp])
    capital_social.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio/capital_social.csv",index=False)

    reserva = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio/reserva.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Passivo Total | Patrimônio Líquido Consolidado | Reservas de Capital"]]
    reserva = pd.concat([reserva,temp])
    reserva.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio/reserva.csv",index=False)

    reserva_reavaliacao = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio/reserva_reavaliacao.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Passivo Total | Patrimônio Líquido Consolidado | Reservas de Reavaliação"]]
    reserva_reavaliacao = pd.concat([reserva_reavaliacao,temp])
    reserva_reavaliacao.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio/reserva_reavaliacao.csv",index=False)

    reserva_lucro = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio/reserva_lucro.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Passivo Total | Patrimônio Líquido Consolidado | Reservas de Reavaliação"]]
    reserva_lucro = pd.concat([reserva_lucro,temp])
    reserva_lucro.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio/reserva_lucro.csv",index=False)

    lucro_prejuizo = pd.read_csv("//home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio/lucro_prejuizo.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Passivo Total | Patrimônio Líquido Consolidado | Lucros/Prejuízos Acumulados"]]
    lucro_prejuizo = pd.concat([lucro_prejuizo,temp])
    lucro_prejuizo.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio/lucro_prejuizo.csv",index=False)

    aj_av_patrimonial = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio/aj_av_patrimonial.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Passivo Total | Patrimônio Líquido Consolidado | Ajustes de Avaliação Patrimonial"]]
    aj_av_patrimonial = pd.concat([aj_av_patrimonial,temp])
    aj_av_patrimonial.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio/aj_av_patrimonial.csv",index=False)

    aj_ac_conversao = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio/aj_ac_conversao.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Passivo Total | Patrimônio Líquido Consolidado | Ajustes Acumulados de Conversão"]]
    aj_ac_conversao = pd.concat([aj_ac_conversao,temp])
    aj_ac_conversao.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio/aj_ac_conversao.csv",index=False)

    outros_resultados = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio/outros_resultados.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Passivo Total | Patrimônio Líquido Consolidado | Outros Resultados Abrangentes"]]
    outros_resultados = pd.concat([outros_resultados,temp])
    outros_resultados.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio/outros_resultados.csv",index=False)

    participacoes = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio/participacoes.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Passivo Total | Patrimônio Líquido Consolidado | Participação dos Acionistas Não Controladores"]]
    participacoes = pd.concat([participacoes,temp])
    participacoes.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio/participacoes.csv",index=False)                      
```
