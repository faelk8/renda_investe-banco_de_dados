```python
def ativo_id():
    """
    Retorna o próximo ID para a tabela patrimonio_liquido_consolidados.ativo_total.
    O ID é gerado com base no maior valor já existente na tabela.
    Se a tabela estiver vazia, o ID começará em 1.
    """
    df = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio_liquido_consolidados/patrimonio_liquido_consolidados.csv')
    if df.empty:
        return 1
    else:
        return df['ativo_id'].max() + 1

def inserir_ativo(base, lista_prefixos, lista_sufixos, lista_valores, ID):
    # con = PostgresConnector()
    for nome, descricao, valor in zip(lista_prefixos, lista_sufixos, lista_valores):
        try:
            if base == "Empréstimos e Financiamentos":
                temp = pd.DataFrame(columns=['ativo_id','nome','descricao','valor'])
                temp['ativo_id'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_emprestimo_financiamento_descritivo = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio_liquido_consolidados/emprestimo_financiamento_descritivo.csv')
                df_emprestimo_financiamento_descritivo = pd.concat([df_emprestimo_financiamento_descritivo, temp], ignore_index=True)
                df_emprestimo_financiamento_descritivo.to_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio_liquido_consolidados/emprestimo_financiamento_descritivo.csv',index=False)
                # tabela = "patrimonio_liquido_consolidados.aplicao_financeira"
                # key = "aplicao_financeira_id"

            elif base == "Outras Obrigações":
                temp = pd.DataFrame(columns=['ativo_id','nome','descricao','valor'])
                temp['ativo_id'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_outras_obrigacoes_descritivo = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio_liquido_consolidados/outras_obrigacoes_descritivo.csv')
                df_outras_obrigacoes_descritivo = pd.concat([df_outras_obrigacoes_descritivo, temp], ignore_index=True)
                df_outras_obrigacoes_descritivo.to_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio_liquido_consolidados/outras_obrigacoes_descritivo.csv',index=False)
                # tabela = "patrimonio_liquido_consolidados.contas_receber"
                # key = "contas_receber_id"
           
            elif "Tributos Diferidos":
                temp = pd.DataFrame(columns=['ativo_id','nome','descricao','valor'])
                temp['ativo_id'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_tributos_descritivo = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio_liquido_consolidados/tributos_descritivo.csv')
                df_tributos_descritivo = pd.concat([df_tributos_descritivo, temp], ignore_index=True)
                df_tributos_descritivo.to_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio_liquido_consolidados/tributos_descritivo.csv',index=False)
                # tabela = "patrimonio_liquido_consolidados.tributos_recuperar"
                # key = "tributos_recuperar_id"
            
            elif base == "Provisões":
                temp = pd.DataFrame(columns=['ativo_id','nome','descricao','valor'])
                temp['ativo_id'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_provisoes_descritivo = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio_liquido_consolidados/provisoes_descritivo.csv')
                df_provisoes_descritivo = pd.concat([df_provisoes_descritivo, temp], ignore_index=True)
                df_provisoes_descritivo.to_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio_liquido_consolidados/provisoes_descritivo.csv',index=False)

            elif base == "Passivos sobre Ativos Não-Correntes a Venda e Descontinuados":
                temp = pd.DataFrame(columns=['ativo_id','nome','descricao','valor'])
                temp['ativo_id'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_provisoes = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio_liquido_consolidados/passivos_descritivo.csv')
                df_provisoes = pd.concat([df_provisoes, temp], ignore_index=True)
                df_provisoes.to_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio_liquido_consolidados/passivos_descritivo.csv',index=False)
            
            elif base == "Lucros e Receitas a Apropriar":
                temp = pd.DataFrame(columns=['ativo_id','nome','descricao','valor'])
                temp['ativo_id'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_lucro = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio_liquido_consolidados/lucro_descritivo.csv')
                df_lucro = pd.concat([df_lucro, temp], ignore_index=True)
                df_lucro.to_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio_liquido_consolidados/lucro_descritivo.csv',index=False)

        except Exception as e:
            logger.error("Erro na função inserir_ativo")
            logger.error(f"Erro ao processar item ({nome}, {descricao}, {valor}): {e}")
            break
    # con.close()

def separar_classes(procurar, resultado, ID):
    lista_prefixos = []
    lista_sufixos = []
    lista_valores = []

    caminhos_pos_procurar = []

    # Coletar todos os caminhos abaixo do "procurar"
    for chave in resultado:
        partes = chave.split(" | ")
        indices_procurar = [i for i, p in enumerate(partes) if p == procurar]
        idx = next((i for i in indices_procurar if i + 1 <= len(partes)), None)
        if idx is not None:
            caminho = partes[idx + 1:]
            caminhos_pos_procurar.append(caminho)

    # Determinar quais caminhos são folhas
    folhas = []
    for caminho in caminhos_pos_procurar:
        is_folha = not any(
            outro != caminho and outro[:len(caminho)] == caminho
            for outro in caminhos_pos_procurar
        )
        if is_folha:
            folhas.append(caminho)

    # Agora, coletar os valores somente dos nós folha
    for chave, valor in resultado.items():
        partes = chave.split(" | ")
        indices_procurar = [i for i, p in enumerate(partes) if p == procurar]
        idx = next((i for i in indices_procurar if i + 1 <= len(partes)), None)
        if idx is None:
            continue

        caminho = partes[idx + 1:]

        if caminho in folhas:
            if len(caminho) == 0:
                lista_prefixos.append(procurar)
                lista_sufixos.append(procurar)
            elif len(caminho) == 1:
                lista_prefixos.append(caminho[0])
                lista_sufixos.append(caminho[0])
            elif len(caminho) > 1:
                lista_prefixos.append(caminho[0])
                lista_sufixos.append(caminho[1])
            lista_valores.append(valor)

    inserir_ativo(procurar, lista_prefixos, lista_sufixos, lista_valores, ID)

def circulante( id=None, COMPANIA_ID=None, dados=None):

    valor_procurado = "Passivo Total | Passivo Não Circulante"
    lista = ["Passivo Total | Passivo Não Circulante | Empréstimos e Financiamentos",
                 "Passivo Total | Passivo Não Circulante | Outras Obrigações",
                 "Passivo Total | Passivo Não Circulante | Tributos Diferidos",
                 "Passivo Total | Passivo Não Circulante | Provisões",
                 "Passivo Total | Passivo Não Circulante | Passivos sobre Ativos Não-Correntes a Venda e Descontinuados",
                 "Passivo Total | Passivo Não Circulante | Lucros e Receitas a Apropriar"
                ]

    resultado_atual = {}

    for chave, valor in dados.items():
        if valor_procurado in chave:
            resultado_atual[chave] = valor

    for l in lista:
        valor_procurado = l
        partes = valor_procurado.split(" | ")
        if len(partes) >= 3:
            procurar = partes[2]

        resultado = {}

        for chave, valor in resultado_atual.items():
            if l in chave:
                resultado[chave] = valor

        separar_classes(procurar, resultado, id)

def inserir_resumo(dados,id):   
    emprestimos = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio_liquido_consolidados/emprestimo_financiamento.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Passivo Total | Passivo Não Circulante | Empréstimos e Financiamentos"]]
    emprestimos = pd.concat([emprestimos,temp])
    emprestimos.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio_liquido_consolidados/emprestimo_financiamento.csv",index=False)

    obrigacoes = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio_liquido_consolidados/outras_obrigacoes.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Passivo Total | Passivo Não Circulante | Outras Obrigações"]]
    obrigacoes = pd.concat([obrigacoes,temp])
    obrigacoes.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio_liquido_consolidados/outras_obrigacoes.csv",index=False)

    tributos = pd.read_csv("//home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio_liquido_consolidados/tributos.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Passivo Total | Passivo Não Circulante | Tributos Diferidos"]]
    tributos = pd.concat([tributos,temp])
    tributos.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio_liquido_consolidados/tributos.csv",index=False)

    provisoes = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio_liquido_consolidados/provisoes.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Passivo Total | Passivo Não Circulante | Provisões"]]
    provisoes = pd.concat([provisoes,temp])
    provisoes.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio_liquido_consolidados/provisoes.csv",index=False)

    passivo = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio_liquido_consolidados/passivos.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Passivo Total | Passivo Não Circulante | Passivos sobre Ativos Não-Correntes a Venda e Descontinuados"]]
    passivo = pd.concat([passivo,temp])
    passivo.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio_liquido_consolidados/passivos.csv",index=False)

    lucro = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio_liquido_consolidados/lucro.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Passivo Total | Passivo Não Circulante | Lucros e Receitas a Apropriar"]]
    lucro = pd.concat([lucro,temp])
    lucro.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/patrimonio_liquido_consolidados/lucro.csv",index=False)
                      
```
