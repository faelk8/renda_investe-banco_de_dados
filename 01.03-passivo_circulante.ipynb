```python
def ativo_id():
    """
    Retorna o próximo ID para a tabela passivo_circulante.ativo_total.
    O ID é gerado com base no maior valor já existente na tabela.
    Se a tabela estiver vazia, o ID começará em 1.
    """
    df = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/passivo_circulante/passivo_circulante.csv')
    if df.empty:
        return 1
    else:
        return df['ativo_id'].max() + 1

def inserir_ativo(base, lista_prefixos, lista_sufixos, lista_valores, ID):
    # con = PostgresConnector()
    for nome, descricao, valor in zip(lista_prefixos, lista_sufixos, lista_valores):
        try:
            if base == "Obrigações Sociais e Trabalhistas":
                temp = pd.DataFrame(columns=['ativo_id','nome','descricao','valor'])
                temp['ativo_id'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_aplicao_financeira = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/passivo_circulante/obr_sociais_trabalhista_descritivo.csv')
                df_aplicao_financeira = pd.concat([df_aplicao_financeira, temp], ignore_index=True)
                df_aplicao_financeira.to_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/passivo_circulante/obr_sociais_trabalhista_descritivo.csv',index=False)

            elif base == "Fornecedores":
                temp = pd.DataFrame(columns=['ativo_id','nome','descricao','valor'])
                temp['ativo_id'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_contas_receber = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/passivo_circulante/fornecedores_descritivo.csv')
                df_contas_receber = pd.concat([df_contas_receber, temp], ignore_index=True)
                df_contas_receber.to_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/passivo_circulante/fornecedores_descritivo.csv',index=False)
            
            elif base == "Obrigações Fiscais":
                temp = pd.DataFrame(columns=['ativo_id','nome','descricao','valor'])
                temp['ativo_id'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_tributos_recuperar = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/passivo_circulante/obr_fiscais_descritivo.csv')
                df_tributos_recuperar = pd.concat([df_tributos_recuperar, temp], ignore_index=True)
                df_tributos_recuperar.to_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/passivo_circulante/obr_fiscais_descritivo.csv',index=False)

            elif base == "Empréstimos e Financiamentos":
                temp = pd.DataFrame(columns=['ativo_id','nome','descricao','valor'])
                temp['ativo_id'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_tributos_recuperar = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/passivo_circulante/emprestimos_descritivo.csv')
                df_tributos_recuperar = pd.concat([df_tributos_recuperar, temp], ignore_index=True)
                df_tributos_recuperar.to_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/passivo_circulante/emprestimos_descritivo.csv',index=False)   

            elif base == "Outras Obrigações":
                temp = pd.DataFrame(columns=['ativo_id','nome','descricao','valor'])
                temp['ativo_id'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_tributos_recuperar = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/passivo_circulante/outras_obrigacoes_descritivo.csv')
                df_tributos_recuperar = pd.concat([df_tributos_recuperar, temp], ignore_index=True)
                df_tributos_recuperar.to_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/passivo_circulante/outras_obrigacoes_descritivo.csv',index=False)   
 
            elif base == "Provisões":
                temp = pd.DataFrame(columns=['ativo_id','nome','descricao','valor'])
                temp['ativo_id'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_outros_ativos_circulantes = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/passivo_circulante/provisoes_descritivo.csv')
                df_outros_ativos_circulantes = pd.concat([df_outros_ativos_circulantes, temp], ignore_index=True)
                df_outros_ativos_circulantes.to_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/passivo_circulante/provisoes_descritivo.csv',index=False)                # tabela = "passivo_circulante.outros_ativos_circulantes"
                # key = "outros_ativos_circulantes_id"

            elif base == "Passivos sobre Ativos Não-Correntes a Venda e Descontinuados":
                temp = pd.DataFrame(columns=['ativo_id','nome','descricao','valor'])
                temp['ativo_id'] = [ID]
                temp['nome'] = [nome]
                temp['descricao'] = [descricao]
                temp['valor'] = [valor]
                df_tributos_recuperar = pd.read_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/passivo_circulante/passivos_descritivo.csv')
                df_tributos_recuperar = pd.concat([df_tributos_recuperar, temp], ignore_index=True)
                df_tributos_recuperar.to_csv('/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/passivo_circulante/passivos_descritivo.csv',index=False)  

        except Exception as e:
            logger.error("Erro na função inserir_ativo")
            logger.error(f"Erro ao processar item ({nome}, {descricao}, {valor}): {e}")
            break
    # con.close()

def separar_classes(procurar, resultado, ID):
    lista_prefixos = []
    lista_sufixos = []
    lista_valores = []

    caminhos_pos_procurar = []

    # Coletar todos os caminhos abaixo do "procurar"
    for chave in resultado:
        partes = chave.split(" | ")
        indices_procurar = [i for i, p in enumerate(partes) if p == procurar]
        idx = next((i for i in indices_procurar if i + 1 <= len(partes)), None)
        if idx is not None:
            caminho = partes[idx + 1:]
            caminhos_pos_procurar.append(caminho)

    # Determinar quais caminhos são folhas
    folhas = []
    for caminho in caminhos_pos_procurar:
        is_folha = not any(
            outro != caminho and outro[:len(caminho)] == caminho
            for outro in caminhos_pos_procurar
        )
        if is_folha:
            folhas.append(caminho)

    # Agora, coletar os valores somente dos nós folha
    for chave, valor in resultado.items():
        partes = chave.split(" | ")
        indices_procurar = [i for i, p in enumerate(partes) if p == procurar]
        idx = next((i for i in indices_procurar if i + 1 <= len(partes)), None)
        if idx is None:
            continue

        caminho = partes[idx + 1:]

        if caminho in folhas:
            if len(caminho) == 0:
                lista_prefixos.append(procurar)
                lista_sufixos.append(procurar)
            elif len(caminho) == 1:
                lista_prefixos.append(caminho[0])
                lista_sufixos.append(caminho[0])
            elif len(caminho) > 1:
                lista_prefixos.append(caminho[0])
                lista_sufixos.append(caminho[1])
            lista_valores.append(valor)

    inserir_ativo(procurar, lista_prefixos, lista_sufixos, lista_valores, ID)

def circulante( id=None, COMPANIA_ID=None, dados=None):

    valor_procurado = "Passivo Total | Passivo Circulante"
    lista = ["Passivo Total | Passivo Circulante | Obrigações Sociais e Trabalhistas",
                 "Passivo Total | Passivo Circulante | Fornecedores",
                 "Passivo Total | Passivo Circulante | Obrigações Fiscais",
                 "Passivo Total | Passivo Circulante | Empréstimos e Financiamentos",
                 "Passivo Total | Passivo Circulante | Outras Obrigações",
                 "Passivo Total | Passivo Circulante | Provisões",
                 "Passivo Total | Passivo Circulante | Passivos sobre Ativos Não-Correntes a Venda e Descontinuados"
                 ]


    resultado_atual = {}

    for chave, valor in dados.items():
        if valor_procurado in chave:
            resultado_atual[chave] = valor

    for l in lista:
        valor_procurado = l
        partes = valor_procurado.split(" | ")
        if len(partes) >= 3:
            procurar = partes[2]

        resultado = {}

        for chave, valor in resultado_atual.items():
            if l in chave:
                resultado[chave] = valor

        separar_classes(procurar, resultado, id)

def inserir_resumo(dados,id):   
    obr_sociais_trabalhista = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/passivo_circulante/obr_sociais_trabalhista.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Passivo Total | Passivo Circulante | Obrigações Sociais e Trabalhistas"]]
    obr_sociais_trabalhista = pd.concat([obr_sociais_trabalhista,temp])
    obr_sociais_trabalhista.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/passivo_circulante/obr_sociais_trabalhista.csv",index=False)

    fornecedores = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/passivo_circulante/fornecedores.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']['Passivo Total | Passivo Circulante | Fornecedores']]
    fornecedores = pd.concat([fornecedores,temp])
    fornecedores.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/passivo_circulante/fornecedores.csv",index=False)

    obigacoes_fiscais = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/passivo_circulante/obr_fiscais.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Passivo Total | Passivo Circulante | Obrigações Fiscais"]]
    obigacoes_fiscais = pd.concat([obigacoes_fiscais,temp])
    obigacoes_fiscais.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/passivo_circulante/obr_fiscais.csv",index=False)

    emprestimo = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/passivo_circulante/emprestimos.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Passivo Total | Passivo Circulante | Empréstimos e Financiamentos"]]
    emprestimo = pd.concat([emprestimo,temp])
    emprestimo.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/passivo_circulante/emprestimos.csv",index=False)

    outras_obrigacoes = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/passivo_circulante/outras_obrigacoes.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Passivo Total | Passivo Circulante | Outras Obrigações"]]
    outras_obrigacoes = pd.concat([outras_obrigacoes,temp])
    outras_obrigacoes.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/passivo_circulante/outras_obrigacoes.csv",index=False)

    provisoes = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/passivo_circulante/provisoes.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Passivo Total | Passivo Circulante | Provisões"]]
    provisoes = pd.concat([provisoes,temp])
    provisoes.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/passivo_circulante/provisoes.csv",index=False)

    passivos = pd.read_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/passivo_circulante/passivos.csv")
    temp = pd.DataFrame(columns=['ativo_id','valor'])
    temp['ativo_id'] = [id]
    temp['valor'] = [dados['data']["Passivo Total | Passivo Circulante | Passivos sobre Ativos Não-Correntes a Venda e Descontinuados"]]
    passivos = pd.concat([passivos,temp])
    passivos.to_csv("/home/rafael/00-Projetos/Renda_Investe/renda_investe-banco_de_dados/data/bd/passivo_circulante/passivos.csv",index=False)    
    
```
